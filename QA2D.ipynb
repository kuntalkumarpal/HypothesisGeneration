```python
punctuationPattern = ["?",".",";",",","!","$","&",")","(",":","@","#"]
```

### Generate conllu and txt files

```python
trainQ = trainData.stem.values
```

```python
atrainA = trainData.Atext.values
btrainA = trainData.Btext.values
ctrainA = trainData.Ctext.values
dtrainA = trainData.Dtext.fillna(" ").values
```

```python
file = "output/train.conllu"
tr = open(file,"w")
filetxt = "output/train.txt"
trtxt = open(filetxt,"w")
```

```python
for num,data in tqdm(enumerate(zip(trainQ,atrainA,btrainA,ctrainA,dtrainA))):
    
    q,a,b,c,d = data[0],data[1],data[2],data[3],data[4]
#     if q[-1] in punctuationPattern: q = q[0:len(q)-1]+" "+q[-1]
#     if a[-1] in punctuationPattern: a = a[0:len(a)-1]+" "+a[-1]
#     if b[-1] in punctuationPattern: b = b[0:len(b)-1]+" "+b[-1]
#     if c[-1] in punctuationPattern: c = c[0:len(c)-1]+" "+c[-1]
#     if d[-1] in punctuationPattern: d = d[0:len(d)-1]+" "+d[-1]
    #print (q)
    q = q.split()
    a = a.split()
    b = b.split()
    c = c.split()
    d = d.split()
    qq,aa,bb,cc,dd = [],[],[],[],[]
    for i in q:
        if i[-1] in punctuationPattern:
            qq.append(i[:-1])
            qq.append(i[-1])
        else: 
            if i[-2:] in "'s":
                qq.append(i[:-2])
                qq.append(i[-2:])
            else:
                qq.append(i)
                
    
    for i in a:
        if i[-1] in punctuationPattern:
            aa.append(i[:-1])
            aa.append(i[-1])
        else: 
            if i[-2:] in "'s":
                aa.append(i[:-2])
                aa.append(i[-2:])
            else:
                aa.append(i)
                
    
    for i in b:
        if i[-1] in punctuationPattern:
            bb.append(i[:-1])
            bb.append(i[-1])
        else: 
            if i[-2:] in "'s":
                bb.append(i[:-2])
                bb.append(i[-2:])
            else:
                bb.append(i)
                
    
    for i in c:
        if i[-1] in punctuationPattern:
            cc.append(i[:-1])
            cc.append(i[-1])
        else: 
            if i[-2:] in "'s":
                cc.append(i[:-2])
                cc.append(i[-2:])
            else:
                cc.append(i)
    
    for i in d:
        if i[-1] in punctuationPattern:
            dd.append(i[:-1])
            dd.append(i[-1])
        else: 
            if i[-2:] in "'s":
                dd.append(i[:-2])
                dd.append(i[-2:])
            else:
                dd.append(i)
                          
    
    #print (qq)
    #print (q,a,b,c,d)
    qstr = ''
    for n,eachWord in enumerate(qq):
        qstr+=str(n+1)+"\t"+eachWord+"\t_"*8+"\n"
    astr = ''
    for n,eachWord in enumerate(aa):
        astr+=str(n+1)+"\t"+eachWord+"\t_"*8+"\n"
    bstr = ''
    for n,eachWord in enumerate(bb):
        bstr+=str(n+1)+"\t"+eachWord+"\t_"*8+"\n"
    cstr = ''
    for n,eachWord in enumerate(cc):
        cstr+=str(n+1)+"\t"+eachWord+"\t_"*8+"\n"
    dstr = ''
    for n,eachWord in enumerate(dd):
        dstr+=str(n+1)+"\t"+eachWord+"\t_"*8+"\n" 
    
    tr.write(qstr+"\n"+astr+"\n"+qstr+"\n"+bstr+"\n"+qstr+"\n"+cstr+"\n"+qstr+"\n"+dstr+"\n")

    #input(" ")
    
    #### ------------ Generate Text File ----------- ####
    
    trtxt.write(data[0]+"\n\n"+data[1]+"\n\n"+data[0]+"\n\n"+data[2]+"\n\n"+data[0]+"\n\n"+data[3]+"\n\n"+data[0]+"\n\n"+data[4]+"\n\n")
    
```

```python
tr.close()
trtxt.close()
```
